# Merge, Join, and Concatenate in Pandas

## Merge

```python
df1 = pd.read_csv(r"C:\Users\AhmedSharaf\Documents\PandasTutorial\LOTR.csv") #first data frame
df1
```

```python
df2 = pd.read_csv(r"C:\Users\AhmedSharaf\Documents\PandasTutorial\LOTR 2.csv")
df2
```

```python
df1.merge(df2) #merge df1 as the left data frame with df2 as the right data frame (inner merge as the default)
```

```python
df1.merge(df2, how = 'inner', on = 'FellowshipID') #inner merge on FellowshipID as the key (notice the two FirstName columns)
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName']) #inner merge on FellowshipID and FirstName (Defaul behaviour)
```

```python
df1.merge(df2, how = 'outer') 
#Outer merge gives back all data whether there is an overlap or not without duplication (notice NaN for missing data)
```

```python
df1.merge(df2, how = 'left')
#Left merge gives back all the data from the left data frame and only the matching data from right data frame
```

```python
df1.merge(df2, how = 'right')
#Right merge gives back all the data from the Right data frame and only the matching data from left data frame
```

```python
df1.merge(df2, how = 'cross')
#Takes every value from the left and compares it to every value from the right data frame 
```

## Join

```python
df1.join(df2, on='FellowshipID', how = 'outer', lsuffix = '_left', rsuffix = '_right')
#Outer join date frame one and two 
#Join is better with indexes
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'),how = 'outer', lsuffix = '_left', rsuffix = '_right')
#Joining both data frames but first we set the indexes for both to FellowshipID
df4
```

## Concatenate

```python
pd.concat([df1,df2])
#concatenate both data frames (vertically add the data from the df2 under df1)
```

```python
pd.concat([df1,df2], join = 'inner')
#concatenat both data frames but only giving the columns that exist in both data frames
```

```python
pd.concat([df1, df2], join = 'outer')
#concatenate both data frames (default)
```

```python
pd.concat([df1, df2], join = 'outer', axis = 1)
#concatenate the data frames on the horizontally (like merge)
```

# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\AhmedSharaf\Documents\PandasTutorial\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates() #Dropping the duplicates in the data
df
```

```python
df = df.drop(columns = "Not_Useful_Column") #Dropping a column
# df.drop(columns = "Not_Useful_Column", inplace = True) #Does the same thing as assinging it to the orginal variable
df
```

```python
# df["Last_Name"] = df["Last_Name"].str.lstrip("...")
# df["Last_Name"] = df["Last_Name"].str.lstrip("/")
# df["Last_Name"] = df["Last_Name"].str.rstrip("_")
#Cleaning "Last_Name" column from unwanted characters
# df
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
#Does the same as the previous cell in one step using Regular Expressions
df
```

*There is a problem with the Phone_Number column data type that made some of the numbers appear as NaN while existing in the original excel file. You might want to check the comments on the video for more information about it.

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("[^a-zA-Z0-9]", '', regex = True)
#Removing any characters that aren't a letter or a number
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x)) #Looping through the column changing every value to a string
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) #Putting - after every 3 numbers
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '') #Replacing nan-- with nothing
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '') #Replacing Na-- with nothing
df
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split("," ,n = 2, expand=True) 
#Spliting the Address column to 3 new columns using a comma as a delimiter and appending them to the data frame
#Had to put (n=2) in newer versions of Pandas
#expand = True shows the columns instead of lists
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
#Replacing every Yes in the Paying Customer column with Y
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N')
#Replacing every No in the Paying Customer column with N
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y')
#Replacing every Yes in the Do_Not_Contact column with Y
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N')
#Replacing every No in the Do_Not_Contact column with N
df
```

```python
df = df.replace('N/a', '') #Replacing every N/a in the data frame with an empty string
df = df.fillna('') #filling evere NaN in the data frame with an empty string
df
```

```python
#Looping through Do_Not_Contact if the value equals 'Y' we drop the intire row to protect the data 
for x in df.index: 
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace = True)
df
```

```python
#Looping through Phone_Number if it's empty we drop the intire row
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)
df
# Another way to drop null values (empty cells in the Phone_Number columns)
#df = df.dropna(subset="Phone_Number", inplace=True)
```
